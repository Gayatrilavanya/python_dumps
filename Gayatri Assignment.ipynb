```
left = pd.DataFrame({'State': ['Missouri', 'Kansas', 'Texas', 'Maryland','Newyork'],
                     'city': ['st.louis', 'wichita', 'dallas', 'baltimore','Buffalo'],
                     'county': ['manchester', 'overlandpark', 'prosper', 'Fredrick','Niagara']})
   
right = pd.DataFrame({'State': ['Missouri', 'Kansas', 'Texas', 'Maryland','Newyork'],
                          'Attractions': ['Meramic', 'skydiving', 'nasa', 'museum','timessquare'],
                          'indianpopulation': ['27000', '30000', '100000', '50000','70000']}) 
```

```
left
```

```
right
```

```
pd.merge(left,right,how='inner',on='State')
```

```
pd.merge(left,right,how='left',on='State')
```

```
pd.merge(left,right,how='right',on='State')
```

```
pd.merge(left,right,how='outer',on='State')
```
