```
import pandas as pd
data={'Company':['NIKE','PUMA','MAC','MAC','BARMIN','NIKE'],
      'Product':['Shoe','socks','lipstick','foundation','eyeliner','shoe'],'Sale':[200,120,340,124,243,350]}
```

```
df = pd.DataFrame(data)
```

```
display(df)
```

```
bycomp=df.groupby('Company')
```

```
bycomp.describe()
```

```
byprod=df.groupby('Product')
```

```
byprod.describe()
```

```
bycomp.std()
```

```
byprod.mean()
```

```
bycomp.describe().transpose()['MAC']
```

```
byprod.describe().transpose()['shoe']
```
