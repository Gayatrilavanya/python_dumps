```
import seaborn as sns
%matplotlib inline
```

```
sns.get_dataset_names()
```

```
diamonds=sns.load_dataset('diamonds')
diamonds.head()
```

```
sns.barplot(x='carat',y='clarity',data=diamonds)
```

```
import numpy as np
```

```
sns.barplot(x='carat',y='clarity',data=diamonds,estimator=np.min)
```

```
sns.barplot(x='carat',y='clarity',data=diamonds,estimator=np.max)
```

```
sns.countplot(x='carat',data=diamonds)
```

```
sns.boxplot(x="clarity", y="price", data=diamonds,palette='rainbow')
```

```
sns.boxplot(data=diamonds,palette='rainbow',orient='h')
```

```
sns.boxplot(x="clarity", y="price", hue="cut",data=diamonds, palette="rainbow")
```

```
sns.violinplot(x="clarity", y="price", data=diamonds,palette='rainbow')
```

```
sns.violinplot(x="cut", y="price", data=diamonds,hue='clarity',palette='Set1')
```

```
sns.stripplot(x="cut", y="price", data=diamonds)
```

```
sns.stripplot(x="clarity", y="price", data=diamonds,jitter=True,hue='cut',palette='Set1')
```

```

sns.factorplot(x='cut',y='price',data=diamonds,kind='bar')
```
